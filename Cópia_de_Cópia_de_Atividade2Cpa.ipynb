Selection Sort

```python
def selectionSort(vetor, n):
    for i in range(1, n):
      min = i

      for j in range(i + 1, n):
        if vetor[j] < vetor[min]:
          min = j
      if i != min:
        aux = vetor[i]
        vetor[i] = vetor[min]
        vetor[min] = aux



    return vetor
```

InsertionSort

```python
def insertionSort(vetor , n):
  for j in range(1, n):
    chave = vetor[j]
    i = j -1

    while i >= 0 and vetor[i] > chave:
      vetor[i+1] = vetor[i]
      i = i -1
    vetor[i+1] = chave

  return vetor
```

**MergeSort** 

```python
class MERGE:
  def __init__(self, contador):
    self.contador = contador;

  def merge(self, vetor, esq, meio, dir):
      n1 = meio - esq + 1;
      n2 = dir - meio;
  
      aux1 = [0] * (n1);
      aux2 = [0] * (n2);
      
      for i in range(0, n1):
          aux1[i] = vetor[esq + i];
  
      for j in range(0, n2):
          aux2[j] = vetor[meio + 1 + j];
  
      i = 0;
      j = 0;
      k = esq;
      while i < n1 and j < n2:
          if aux1[i] <= aux2[j]:
              vetor[k] = aux1[i];
              i += 1;
              self.contador += 1;
          else:
              vetor[k] = aux2[j];
              j += 1;
              self.contador += 1;
          k += 1;
          self.contador += 1;
      
      while i < n1:
          vetor[k] = aux1[i];
          i += 1;
          k += 1;
          self.contador += 1;
      while j < n2:
          vetor[k] = aux2[j];
          j += 1;
          k += 1;
          self.contador += 1;

  def mergeSort(self,vetor, esq, dir):
      if esq < dir:    
          meio = esq+(dir-esq)//2;
          self.mergeSort(vetor, esq, meio);
          self.mergeSort(vetor, meio+1, dir);
          self.merge(vetor, esq, meio, dir);

      return vetor;

  def getContador2(self):
      return self.contador;
```

**QuickSort**

```python
class QUICK:

  def __init__(self, contador):
    self.contador = contador;

  def particao(self,vetor, esq, dir):
        pivo = vetor[dir];
        i = esq -1;
        for j in range(esq, dir):
          if vetor[j] <= pivo:
            i += 1;
            aux = vetor[i];
            vetor[i] = vetor[j];
            vetor[j] = aux;
            self.contador+=1;

        aux = vetor[i+1];
        vetor[i+1] = vetor[dir];
        vetor[dir] = aux;
        self.contador+=1
        return i + 1;

  def quickSort(self,vetor, esq, dir):

      if esq < dir:
          pivo = self.particao(vetor ,esq , dir);
          self.quickSort(vetor ,esq , pivo-1);
          self.quickSort(vetor, pivo + 1 , dir);
      return vetor;

  def getContador(self):
          return self.contador;
```

QuickSort 

```python
class Quick:
  def __init__(self, contador):
    self.contador = contador;
  
  def particao(self,vetor, esq, dir):
      pivo = vetor[dir];
      i = esq -1;

      for j in range(esq, dir):
        if vetor[j] <= pivo:
          i += 1;
          aux = vetor[i];
          vetor[i] = vetor[j];
          vetor[j] = aux;
          self.contador += 1;

      aux = vetor[i+1];
      vetor[i+1] = vetor[dir];
      vetor[dir] = aux;
      self.contador += 1;
      
      return i + 1;

  def quickSort(self,vetor, esq, dir):
      if esq < dir:
        pivo = self.particao(vetor , esq , dir);
        self.quickSort(vetor , esq , pivo -1);
        self.quickSort(vetor, pivo + 1 , dir);
    
      return vetor;

  def getContador(self):
      return self.contador;
  
```

QuickSort com o pivo aleatorio

```python
import numpy as np
import sys
sys.setrecursionlimit(1500)

class QUICKALE:
    def __init__(self, contador):
      self.contador = contador;

    def swap(self,vetor, i , j):
        aux = vetor[i];
        vetor[i] = vetor[j];
        vetor[j] = aux;
        self.contador += 1;


    def particao(self,vetor, esq, dir):

        rand = np.random.randint(dir-esq+1) + esq;
        self.swap(vetor , rand, dir);
        pivo = vetor[dir];
        i = esq-1;

        for j in range(esq, dir):
          if vetor[j] <= pivo:
            i += 1;
            self.swap(vetor, i, j);
            self.contador += 1;

        self.swap(vetor, i+1, dir);
        
        return i + 1;

    def quickSortA(self,vetor, esq, dir):
        if esq < dir:
          pivo = self.particao(vetor , esq , dir);
          self.quickSortA(vetor , esq , pivo -1);
          self.quickSortA(vetor, pivo + 1 , dir);
      
        return vetor;
    
    def getContador(self):
      return self.contador;
```
